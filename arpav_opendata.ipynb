# XML files (che per qualche ragione sono json?)

```r
library(readxl)
library(tidyverse)
library(dplyr)
library(stringr)
library(ggplot2)
library(gridExtra)
library(grid)
library(xml2)
library(jsonlite)
```

```r
stations_file <- fromJSON("https://api.arpa.veneto.it/REST/v1/meteo_storici?anno=2010&coordcd=18")
stations_flt <- stations_file$data %>%
  filter(nome_stazione %in% c('Auronzo', 'Castelfranco Veneto', 'Porto Tolle - Pradon', 'Roverchiara'))
codseq <- as.vector(stations_flt$codseq)
names(codseq)<-c('Auronzo', 'Castelfranco', 'PortoTolle', 'Roverchiara')
codseq ### servono per chiamare i file dopo
```

```r
lista_dati <- list()

for (anno in 2010:2025) {
    for (i in 1:4) {
        url <- paste0("https://api.arpa.veneto.it/REST/v1/meteo_storici_tabella?anno=", 
                     anno, "&codseq=", codseq[i])
        
        df <- fromJSON(url)$data %>%
        mutate(min = sapply(valore, function(x) fromJSON(x)$MINIMO),
            media = sapply(valore, function(x) fromJSON(x)$MEDIO),
            max = sapply(valore, function(x) fromJSON(x)$MASSIMO),
            data = as.Date(dataora)) %>%
            select(nome_stazione, data, min, media, max)

        lista_dati[[length(lista_dati) + 1]] <- df
    }
}

# Combina tutti i dataframe
final_df <- bind_rows(lista_dati)
```

```r
final_df ## alcuni giorni mancano, ma mancano proprio nel dataset
write.csv(final_df,"dati_arpa.csv", row.names = FALSE)
```
